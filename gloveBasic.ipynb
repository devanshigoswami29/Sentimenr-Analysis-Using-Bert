```python
import numpy as np

words = dict()

def add_to_dict(d,data):
  with open(data, 'r')as f :
    for line in f.readlines():
      line = line.split(' ')

      try:
        d[line[0]] = np.array(line[1:], dtype = float)
      except:
        continue
```
